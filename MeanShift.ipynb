```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from mpl_toolkits.mplot3d import Axes3D
```

```python
sns.set_style("whitegrid")
```

```python
centers_defined = [[2,2,2],[7,7,7],[5,13,13]]
```

```python
X,_ = make_blobs(n_samples=150,centers=centers_defined,cluster_std=0.6)
```

```python
X
```

```python
fig = plt.figure(figsize=(8,12))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2])
plt.show()
```

```python
# @title MeanShift
```

```python
from sklearn.cluster import MeanShift
```

```python
ms = MeanShift()
ms.fit(X)
```

```python
cluster_pred = ms.fit_predict(X)
```

```python
cluster_pred
```

```python
fig = plt.figure(figsize=(8,12))
ax = fig.add_subplot(111,projection ="3d")
ax.scatter(X[:,0],X[:,1],X[:,2],cmap="rainbow", c=cluster_pred)
plt.show()
```
