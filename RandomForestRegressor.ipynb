```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,1].values
```

```python
X
```

```python
X.shape
```

```python
X = X.reshape(-1,1)
```

```python
X.shape
```

```python
X
```

```python
y = dataset.iloc[:,2].values
```

```python
y
```

```python
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=4)
```

```python
model.fit(X,y)
```

```python
model.predict(np.array([6.5]).reshape(-1,1))
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
model.score(X,y)
```
